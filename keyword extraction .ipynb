```python
file_names=os.listdir("/Users/anees/OneDrive/Documents/sem6/NLP/SemEval2017/docsutf8")
texts = []
keys = []

file_path =" /Users/anees/OneDrive/Documents/sem6/NLP/SemEval2017/docsutf8/"
key_path = "/Users/anees/OneDrive/Documents/sem6/NLP/SemEval2017/keys/"

for file_name in file_names:
    key_name = file_name.split(".")[0] + ".key"
    
    with open(os.path.join(docs_utf,file_name), encoding="utf8") as f:
        text = f.read()
        texts.append(text)
        
    with open(os.path.join(key_path,key_name), encoding="utf8") as f:
        key = f.readlines()
        key = [txt[:-1] for txt in key]
        key = ", ".join(key)
        keys.append(key)
     
```

```python
import spacy 
NER=spacy.load("en_core_web_sm")
```

```python
def ner_function(text):
    doc=NER(text)
    entities=[ent.text for ent in doc.ents if ent.label_ in['PERSON','ORG',"GPE"]]
    return entities
```

```python
def pos_function(text):
    doc=NER(text)
    pos_tags=[token.text+"_"+token.pos_ for token in doc]
    
    return pos_tags
```

```python
def combined(text):
    entities=ner_function(text)
    pos_tags=pos_function(text)
    combined=entities+pos_tags
    return entities
```

```python
from rake_nltk import Rake 
def extract_keywords(text):
    r=Rake()
    r.extract_keywords_from_text(text)
    keywords=r.get_ranked_phrases()
    return keywords
```

```python
df['keywords_rake']=df["Text"].apply(extract_keywords)
```

```python

pos_and_ner=[]
for text in df['Text']:
    pos_ner=combined(text)
    pos_and_ner.append(pos_ner)
```

```python
df['key_pos_ner']=pos_and_ner
```
